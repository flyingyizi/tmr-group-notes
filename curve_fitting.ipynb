# source question

```python
import numpy as np

from matplotlib import pyplot as plt

xs = np.array([1., 2.22,  3.44,  4.66,  5.88, 7.11,  8.33,  9.55, 10.77, 12])
ys = np.array([0.05, 0.14, 0.29, 0.34, 0.51, 0.6, 0.63, 0.65, 0.87, 0.88])

def f(x):
  return x * 0.05

plt.plot(xs, f(xs))
plt.scatter(xs, ys)
```

# 简单线性回归

使用回归分析绘制拟合曲线是一种常见的方法，简单线性回归就是其中的一种。简单线性回归可以通过最小二乘法来计算回归系数。

```python
# 原始数据
plt.scatter(xs, ys)
# 计算回归系数
slope, intercept = np.polyfit(xs,ys,1)
#绘制拟合曲线
plt.plot(xs, slope*xs + intercept, color='red')
```

# 多项式回归

使用多项式回归是一种常用方法，它可以用来拟合更加复杂的数据集。

```python
# 原始数据
plt.scatter(xs, ys)
# 计算多项式回归系数
coefs = np.polyfit(xs,ys,3)
#生成多项式拟合对象
poly = np.poly1d(coefs)
#生成新的横坐标，使得拟合更加平滑
new_xs = np.linspace(min(xs),max(xs),1000)
plt.plot(new_xs, poly(new_xs),color='red')
```
